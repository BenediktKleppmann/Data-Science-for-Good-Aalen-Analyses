## Check which sensors are in BW

### 2.0 get .poly-file
downloaded from https://download.geofabrik.de/europe/germany/baden-wuerttemberg.html 

### 2.1  Convert .poly-file to .shp-file

```python
import pandas as pd
from shapely.geometry import mapping, Polygon
import fiona

poly_list = []
bw_df = pd.read_csv('baden-wuerttemberg.poly')
print(bw_df)
for index, row in bw_df.iterrows():
    poly_list.append((row['lat'], row['long']))


poly = Polygon(poly_list)


# Define a polygon feature geometry with one attribute
schema = {
    'geometry': 'Polygon',
    'properties': {'id': 'int'},
}

# Write a new Shapefile
with fiona.open('baden-wuerttemberg.shp', 'w', 'ESRI Shapefile', schema) as c:
    ## If there are multiple geometries, put the "for" loop here
    c.write({
        'geometry': mapping(poly),
        'properties': {'id': 123},
    })
```

### 2.2 check which sensors are in BW or Ostalbkreis

```python
import shapefile
from shapely.geometry import shape, Point
import pandas as pd
r_bw = shapefile.Reader("baden-wuerttemberg.shp")
shapes_bw = r_bw.shapes()
polygon_bw = shape(shapes_bw[0])  

r_oak = shapefile.Reader("baden-wuerttemberg.shp")
shapes_oak = r_oak.shapes()
polygon_oak = shape(shapes_oak[0])  

sensor_locations_df = pd.read_csv('sensor_locations.csv')
sensor_locations_df['in_bw'] = None
sensor_locations_df['in_ostalbkreis'] = None
for index, row in sensor_locations_df.iterrows():

    point = Point(row['longitude'], row['latitude'])
    sensor_locations_df.loc[index, 'in_bw'] = polygon_bw.contains(point)
    sensor_locations_df.loc[index, 'in_ostalbkreis'] = polygon_oak.contains(point)
sensor_locations_df.to_csv('data/sensor_locations.csv', index=False)
```
